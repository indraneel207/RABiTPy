## Additional Examples

### Dual Channel

In this example, we will focus on videos with dual channels. Instead of identifying only grayscale objects, we aim to detect objects that appear in the green channel as well as those in the red channel. Once we have completed this process, we can apply our package to achieve the desired tracking, with separate tracks generated for each channel.

Lets take this image frame for example, that has both the green and red channels.

```python
import matplotlib.pyplot as plt
import cv2

# Read the image
image_path = './sample_images/first_frame.tiff'
img = cv2.imread(image_path)

# Convert the image from BGR to RGB
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(RGB_img)
plt.axis('off')
plt.show()
```

Here, if we want to separate the channels, we can use the following code to do so:

```python
from PIL import Image

# Load and Split Image: Red, Green, Blue
image = Image.open(image_path)
r, g, b = image.split()

# Merge the images
r_img = Image.merge("RGB", (r, Image.new("L", r.size), Image.new("L", r.size)))
g_img = Image.merge("RGB", (Image.new("L", r.size), g, Image.new("L", r.size)))
b_img = Image.merge("RGB", (Image.new("L", r.size), Image.new("L", r.size), b))

# Display the images
plt.figure(figsize=(18, 6))

# Red channel image
plt.subplot(1, 3, 1)
plt.imshow(r_img)
plt.title('Red Channel Image')
plt.axis('off')

# Green channel image
plt.subplot(1, 3, 2)
plt.imshow(g_img)
plt.title('Green Channel Image')
plt.axis('off')

# Blue channel image
plt.subplot(1, 3, 3)
plt.imshow(b_img)
plt.title('Blue Channel Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

As shown, the channels are split to isolate different objects based on their colors. The red channel image displays only the red objects, the green channel image shows the green objects, and the blue channel image highlights the blue objects. All other objects appear as black in these images because the other color channels are set to zero. You can repeat this process for other image frames. Once you have the channel-specific images, you can use them to detect the objects of interest using the package.
