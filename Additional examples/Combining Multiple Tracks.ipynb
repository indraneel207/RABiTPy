### Additional Examples: Working with Multiple Tracks and Combining Them to Obtain MSD Stats

This example demonstrates how to work with multiple track data, combine them into a single Tracker object, and compute Mean Squared Displacement (MSD) statistics. This process is useful for analyzing the movement and diffusion characteristics of particles over time.

```python
from microbe_vision.capture import Capture
from microbe_vision.identify import Identify
from microbe_vision.track import Tracker
from microbe_vision.utils import Utility

from typing import List

# Matplotlib settings for better visualization
%matplotlib inline
```

```python

# List of input CSV files containing particle tracking data
files = ['../input_files/linked_particles.csv']

# Combine the CSV files into a single Tracker object
# Parameters:
# - working_directory: Directory where input files are located
# - pixel_scale_factor: Scale factor to convert pixels to micrometers
# - scale_units: Units for the scale (e.g., micrometers)
# - capture_speed_in_fps: Speed of capture in frames per second
track = Utility.combine_csvs_to_tracker(
    files,
    working_directory='input_files',
    pixel_scale_factor=0.1666,
    scale_units='um',
    capture_speed_in_fps=15
)
```

```python
# Sort and plot scatter plots of the particle trajectories
track.sort_and_plot_scatter_of_trajectories()
```

```python
# Compute the Mean Squared Displacement (MSD) for the combined tracks
# MSD analysis helps in understanding the diffusion and movement patterns of particles
# Parameters:
# - max_lag_time: Maximum lag time for MSD calculation
track.compute_plot_save_MSD(max_lag_time=100)
```

- **`combine_csvs_to_tracker`**: This function reads particle tracking data from CSV files and combines it into a single `Tracker` object. You can specify the pixel scale factor to convert pixel measurements to real-world units (micrometers) and the capture speed in frames per second.

- **`sort_and_plot_scatter_of_trajectories`**: This method sorts the trajectories of particles and creates scatter plots, which helps visualize the paths of the particles.

- **`compute_plot_save_MSD`**: This method calculates and plots the Mean Squared Displacement (MSD) of the particles. MSD is a common statistical measure used to analyze the movement and diffusion of particles over time. The `max_lag_time` parameter specifies the maximum time lag for which MSD will be calculated.
